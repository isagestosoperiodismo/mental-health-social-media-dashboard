```python
df = df.drop_duplicates()

df["Gender"] = df["Gender"].str.strip().str.title()
df["Stress_Level"] = df["Stress_Level"].str.strip().str.title()

print(df["Age"].min(), df["Age"].max())
print(df["Sleep_Hours_Per_Night"].min(), df["Sleep_Hours_Per_Night"].max())
print(df["Avg_Daily_Usage_Hours"].min(), df["Avg_Daily_Usage_Hours"].max())
```

```python
def clasificar_uso(horas):
    if horas <= 2:
        return "0-2h (Bajo)"
    elif horas <= 5:
        return "2-5h (Medio)"
    else:
        return "5h+ (Alto)"

df["Rango_Uso_Diario"] = df["Avg_Daily_Usage_Hours"].apply(clasificar_uso)
```

```python
df["Indice_Riesgo"] = (df["Avg_Daily_Usage_Hours"] + df["Daily_Unlocks"]/10) / (df["Physical_Activity_Hours"] + 0.1)
```
